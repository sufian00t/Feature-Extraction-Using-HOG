```python
normal_path = '/home/rupok1/Downloads/Normal'
sick_path = '/home/rupok1/Downloads/Sick'
```

```python
normal_images = []
for image_file in os.listdir(normal_path):
    image_path = os.path.join(normal_path, image_file)
    image = cv2.imread(image_path)
    
    if image is not None:
        normal_images.append(image)
```

```python
normal_images
```

```python
sick_images = []
for image_file in os.listdir(sick_path):
    image_path = os.path.join(sick_path, image_file)
    image = cv2.imread(image_path)
    
    if image is not None:
        sick_images.append(image)
```

```python
sick_images
```

```python
def preprocess_image(image):
    image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    image = cv2.resize(image, (224, 224))
    image = image / 255.0
    return image
```

```python
preprocessed_normal_images = []
for image in normal_images:
    preprocessed_image = preprocess_image(image)
    preprocessed_normal_images.append(preprocessed_image)
```

```python
preprocessed_normal_images
```

```python
preprocessed_sick_images = []
for image in sick_images:
    preprocessed_image = preprocess_image(image)
    preprocessed_sick_images.append(preprocessed_image)
```

```python
preprocessed_sick_images
```

```python
def extract_hog_features(image):
    hog_features = hog(image, orientations=8, pixels_per_cell=(8, 8), cells_per_block=(2, 2))
    return hog_features
```

```python
normal_hog_features = []
for image in preprocessed_normal_images:
    hog_features = extract_hog_features(image)
    normal_hog_features.append(hog_features)
```

```python
normal_hog_features
```

```python
sick_hog_features = []
for image in preprocessed_sick_images:
    hog_features = extract_hog_features(image)
    sick_hog_features.append(hog_features)
```

```python
sick_hog_features
```

```python
all_hog_features = np.concatenate((normal_hog_features, sick_hog_features))
```

```python
num_features = all_hog_features.shape[1]
column_names = ["feature_" + str(i) for i in range(num_features)]

hog_df = pd.DataFrame(all_hog_features, columns=column_names)
```

```python
hog_df
```

```python
normal_targets = np.zeros(len(normal_hog_features), dtype=int)
sick_targets = np.ones(len(sick_hog_features), dtype=int)
```

```python
all_targets = np.concatenate((normal_targets, sick_targets))
```

```python
hog_df = hog_df.assign(new_column=all_targets)
```
